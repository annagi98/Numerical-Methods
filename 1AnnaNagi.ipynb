```python
#zad 6
from math import sqrt
from numpy import array
from numpy.linalg import inv, det

a = sqrt(2)
M = array([ [a, 1, -a],     #macierz M
            [0, 1, 1],
            [-a, a, 1]])

print("\nMacierz normalna:")
print("{0}".format(M))

macierz_odwrotna = inv(M)
print("\nMacierz odwrotna:")
print("{0}".format(macierz_odwrotna))

macierz_transponowana = M.transpose()
print("\nMacierz transponowana:")
print("{0}".format(macierz_transponowana))

wyznacznik_M = det(M)
print("\nWyznacznik macierzy M = ", wyznacznik_M)



#zad 7 - wypisywanie element√≥w macierzy M

print("M(1,1) = ", M[0, 0])
print("M(3,3) = ", M[2, 2])
print("M(3,2) = ", M[2, 1])


w1 = M[:, 2]                                #wektor w1
print("\n\nWektor w1: \n {0}".format(w1))

w2 = M[1, :]                                #wektor w2
print("\nWektor w2: \n {0}".format(w2))
```

```python
#zad 9

from numpy import arange, linspace

#podajemy poczatek, koniec i krok
#- komputer sam wylicza ilosc licb w ciagu
#w arange liczba koncowa nie nalezy do przedzialu liczb w ciagu,
#a w linspace nalezy
ciag1 = arange(1, 19, 1.5)    
print("\nCiag 1 (arange): ", ciag1)

#podajemy poczatek, koniec i ilosc liczb pomiedzy nimi
#- komputer sam oblicza krok
ciag2 = linspace(1, 19, 13)   
print("\nCiag 2 (linspace): ", ciag2)
```

```python
#zad 10
from numpy import linspace
import matplotlib.pyplot as plt

def funkcja1(x):
    return x**3 - 3*x

#wykres 1.
x = linspace(-1, 1)
y = [funkcja1(i) for i in x]

plt.plot(x, y)
plt.title('Wykres dla x nalezacego do przedzialu od -1 do 1: ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('Wykres 1.')
plt.show()


#wykres 2.
x = linspace(-5, 5)
y = [funkcja1(i) for i in x]

plt.plot(x, y)
plt.title('Wykres dla x nalezacego do przedzialu od -5 do 5: ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('Wykres 2.')
plt.show()


#wykres 3.
x = linspace(0, 5)
y = [funkcja1(i) for i in x]

plt.plot(x, y)
plt.title('Wykres dla x nalezacego do przedzialu od 0 do 5: ')
plt.xlabel('x')
plt.ylabel('y')
plt.legend('Wykres 3.')
plt.show()
```

```python
#zad 10
from numpy import linspace
import matplotlib.pyplot as plt
from matplotlib.pyplot import semilogy


def cieplo(m_g, v_km_h):
    m_kg = m_g / 1000

    v_m_h = v_km_h * 1000
    v_m_s = v_m_h / 3600

    return (m_kg * v_m_s**2) / 2


wynik = cieplo(2500, 60)
print("\nWynik w d≈ºulach: ", wynik)

q_kcal = wynik * 4186.8
print("\nWynik w kilokaloriach: ", q_kcal)


v_km_h = linspace(200, 0)

y = [cieplo(3000, i) for i in v_km_h]

plt.plot(v_km_h, y)
plt.title('Zaleznosc liniowa ilosci ciepla wydzielonego podczas hamowania od predkosci dla ciala o masie 3000g: ')
plt.xlabel('predkosc')
plt.ylabel('ilosc ciepla')
plt.legend('Wykres 1. (skala liniowa) ')
plt.show()


semilogy(v_km_h, y)
plt.title('Zaleznosc logarytmiczna ilosci ciepla wydzielonego podczas hamowania od predkosci dla ciala o masie 3000g: ')
plt.xlabel('predkosc')
plt.ylabel('ilosc ciepla')
plt.legend('Wykres 2. (skala logarytmiczna) ')
plt.show()
```
